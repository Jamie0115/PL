```python
import pandas as pd
from collections import Counter
orgdata = pd.read_csv("Scenic.csv")
orgdata.head()
```

```python
orgdata['seg'] = ""
orgdata.head()
```

```python
subdata = orgdata[['Toldescribe', 'seg']]
```

```python
output_series = subdata.groupby(['Toldescribe'])['seg'].apply(lambda x: ','.join(x)).reset_index()
```

```python
output_series
```

```python
output_series['Toldescribe'] = output_series['Toldescribe'].str.replace("', '", " ")
output_series
```
